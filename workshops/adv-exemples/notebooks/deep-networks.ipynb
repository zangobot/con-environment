# Evading Anti-malware Deep Neural Networks

In this exercise, you will familiarize with libraries to instantiate (and fool) deep neural networks trained to detect Windows malware.

```python
from secmlt.metrics.classification import Accuracy
import torch
from pathlib import Path
from maltorch.data.loader import load_from_folder, create_labels
from maltorch.zoo.malconv import MalConv
from torch.utils.data import TensorDataset, DataLoader

device = "cpu"

deep_net = MalConv().create_model(device=device)
data = load_from_folder(Path("./malware_samples"), device=device, limit=100)
dataloader = DataLoader(TensorDataset(data, create_labels(data, 1, device=device)), batch_size=32)
accuracy = Accuracy()(deep_net, dataloader)
print(f"Accuracy on known malware: {accuracy:.2f}")

```
